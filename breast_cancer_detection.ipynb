importing all dependendcies libraries


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and Analysis

```python
cancer_dataset = pd.read_csv('/content/breast-cancer.csv')
```

```python
cancer_dataset.head()
```

```python
cancer_dataset.shape
```

```python
#converting diagnosis label B as 0 and M as 1
cancer_dataset.replace({"diagnosis": {'B':0, 'M':1}}, inplace=True)
```

```python
#finding statistical approach
cancer_dataset.describe()
```

```python
cancer_dataset['diagnosis'].value_counts()
```

```python
cancer_dataset.groupby('diagnosis').mean()
```

separating data from its label

```python
X = cancer_dataset.drop(columns ='diagnosis', axis=1)
Y = cancer_dataset['diagnosis']
print(Y)
```

Data preprocessig and Analysis

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
print(standardized_data)
```

```python
X = standardized_data
Y = cancer_dataset['diagnosis']
print(Y)
```

TRAIN TEST SPLIT

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

TRAINING THE MODEL

```python
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

MODEL EVALUATION AND ACCURACY SCORE

```python
#accuracy score on training dataset
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on testing data = ', testing_data_accuracy)
```
